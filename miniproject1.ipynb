```python
import random as r

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

for i in range(0,3000):
    sett = r.sample(Set,5)
    if sum(sett)==0:
        print(sett)
```

```python
for i in range(0,3000):
    sett = r.sample(Set,r.randint(3,6))
    if sum(sett)==0:
        print(sett)
```
